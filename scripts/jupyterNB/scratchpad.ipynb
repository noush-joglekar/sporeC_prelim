```python
import pyarrow.parquet as pq
import pandas as pd

f = '/gpfs/commons/groups/gursoy_lab/ajoglekar/Projects/2023_03_01_multiwayInteractions/2023_03_01_v0_dataGathering/PoreC_Fibroblasts/adult/NlaIII_run04_batch2_GRCh38_unphased.at.pore_c.parquet'
cM = '/gpfs/commons/groups/gursoy_lab/ajoglekar/Projects/2023_03_01_multiwayInteractions/2023_03_01_v0_dataGathering/PoreC_Fibroblasts/chr_alias.tbl'

chrMap = pd.read_table(cM).set_index('alias2')['chrom'].to_dict()

df = pq.read_table(source=f).to_pandas()
```

```python
def replace_with_dict(value):
    if value in chrMap:
        return chrMap[value]
    else:
        return None
```

```python
filtered_df = df[df['pass_filter'] == True]
filtered_df.loc[:,'chrom'] = filtered_df['chrom'].apply(replace_with_dict)
filtered_df = filtered_df.dropna(subset=['chrom'])
filtered_df.head()
filtered_df.reset_index(drop=True, inplace=True)

readCounts = filtered_df.groupby('read_idx').size().reset_index(name='count')
filteredRead_ix = readCounts[readCounts['count'] >= 2]


filtered_df = filtered_df[filtered_df['read_idx'].isin(filteredRead_ix['read_idx'])]
```
